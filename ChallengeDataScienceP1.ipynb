# 4 - Join Normalizations

```python
def normalize_json(dataframe):
  return_dataframe = pd.DataFrame()
  for column in list(data.columns[2:]):
    dataframe_column = pd.json_normalize(dataframe[column])
    return_dataframe = pd.concat([return_dataframe, dataframe_column], axis=1)

  return pd.concat([dataframe[list(data.columns[:2])], return_dataframe], axis=1)
```

```python
data = normalize_json(data)
data
```

```python
data.info()
```

# 5 - Verificando inconsistência nos dados

```python
for coluna in data.columns:
    print('---')
    print(data[coluna].value_counts())
```

É possível perceber na variável Churn que há uma categoria que não está nomeada, representando dados faltantes. Devemos remover essa categoria pois não traz dados válidos para a análise

```python
# Verificando quantos dados existem
data['Churn'].value_counts()
```

```python
data = data[data['Churn']!=''] # Pegando elementos vazios de churn
data.reset_index(drop=True, inplace=True)
```

Podemos ver agora não apresenta mais dados vazios

```python
# Verificando quantos dados existem
data['Churn'].value_counts()
```

Outra coluna que apresenta dados vazios (' ') é a Charges.Total. Essa coluna tem uma relação com Charges.Monthly e tenure.



```python
data.query('tenure == 0')[['Charges.Total', 'Charges.Monthly', 'tenure']]
```

```python
# Pegando todas as linhas onde a coluna "Charges.Total" é vazia.

data[data['Charges.Total'] == ' '][['Charges.Total', 'Charges.Monthly', 'tenure']]
```

```python
idx = data[data['Charges.Total'] == ' '].index
data.loc[idx, "Charges.Total"] = data.loc[idx, "Charges.Monthly"]
```

```python
data.query('tenure == 0')[['Charges.Total', 'Charges.Monthly', 'tenure']]
```

Modificando o tipo da variável

```python
data['Charges.Total'] = data['Charges.Total'].astype('float64')
```
